```python
import numpy as np
from scipy.spatial.distance import euclidean
from scipy.spatial.distance import cosine
```

```python
from UBCFBase import UBCFBase
```

```python
import pandas as pd
```

```python
raw_rating_df = pd.read_csv('new_ratings_all.txt',
                                    sep=':',
                                    header=None,
                                    names=['uid', 'iid', 'rating', 'date', 'comment_type'],
                                    dtype={'uid': np.str, 'iid': np.str})
```

```python
uci = UBCFBase.compute_UCI(raw_rating_df)
```

```python
urdi = UBCFBase.compute_URDI(raw_rating_df)
```

```python
behavior_matrix = pd.concat([uci, urdi], axis=1)
```

```python
from sklearn.metrics.pairwise import pairwise_distances
```

```python
beavior_sim = pairwise_distances(behavior_matrix)
```

```python
beavior_sim = pd.DataFrame(beavior_sim, index=behavior_matrix.index, columns=behavior_matrix.index)
```
